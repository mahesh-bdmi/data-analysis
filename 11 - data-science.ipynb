```python
import pandas as pd
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon[pokemon['Type 1']=='Fire']['Name']
```

```python
pokemon.sort_values(['Attack', 'Defense'], ascending=[0,0])
```

```python
pokemon['Total'] = pokemon.iloc[:,4:10].sum(axis=1)
```

```python
cols = list(pokemon.columns)
pokemon = pokemon[cols[0:4]+[cols[-1]]+cols[4:12]] #rearranging the columns
```

```python
pokemon.to_csv('pokemonv2.csv',index=False)
```
