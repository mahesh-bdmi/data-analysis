```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sales = pd.read_csv('data/bikes_sales_data.csv', parse_dates=['Date'])
```

```python
sales.head(10)
```

```python
sales.shape
```

```python
sales.info()
```

```python
sales.describe()
```

```python
sales['Unit_Cost'].plot(kind='box',vert=False,figsize=(14,6))
```

```python
sales['Unit_Cost'].plot(kind='density',figsize=(14,6))
```

```python
dgraph = sales['Unit_Cost'].plot(kind='density',figsize=(14,6))
dgraph.axvline(sales['Unit_Cost'].mean(),color='red')
dgraph.axvline(sales['Unit_Cost'].median(),color='green')
```

# Categorical analysis and visualization
Let's analysis the Age_Group column

```python
sales['Age_Group'].value_counts()
```

```python
sales.Age_Group.value_counts().plot(kind='pie',figsize=(6,6))
```

```python
corr = sales.corr()
corr
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr,cmap='RdBu',fignum=fig.number)
plt.xticks(range(len(corr.columns)),corr.columns,rotation='vertical')
plt.yticks(range(len(corr.columns)),corr.columns)
```
